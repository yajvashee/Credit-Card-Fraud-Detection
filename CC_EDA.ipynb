Check Feature Variance

```python
variance = df.var().sort_values()
variance
```

Every predictor (V1–V28, Amount, and Time) has a variance greater than zero.
Therefore, every feature contains some variation and potentially useful information.

The large variances for Amount and especially Time are expected because:

Amount is measured in currency units and spans a wide range.
Time is measured in seconds over approximately 48 hours (0–172,800 seconds).

The PCA features (V1–V28) have much smaller variances because they are transformed features on a standardized scale.

Pairplot

```python
import seaborn as sns

sample = df.sample(1000)

sns.pairplot(sample[['V1','V2','V3','V4','Class']], hue='Class')
```

A pairplot allows you to examine the relationships between pairs of variables and how well the classes (fraud vs. non-fraud) are separated.

Observation: The pairplot of selected PCA features (V1, V2, V3, and V4) indicates varying degrees of separation between fraudulent and legitimate transactions. Some feature combinations show distinct clustering of the two classes, suggesting that these variables contain discriminative information useful for fraud detection. However, there is also overlap between the classes, indicating that no single feature can perfectly distinguish fraudulent from legitimate transactions. This supports the use of machine learning models that can learn patterns from multiple features simultaneously.

Distribution of Fraud Across Time

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.scatter(
    df["Time"],
    df["Class"],
    alpha=0.3,
    s=10
)

plt.title("Fraud Transactions Over Time")
plt.xlabel("Time")
plt.ylabel("Class")

plt.show()
```

This scatter plot shows the relationship between Time (x-axis) and Class (y-axis). Shows whether fraud occurs uniformly or in bursts.

Class = 0 → Legitimate transactions (bottom row)
Class = 1 → Fraudulent transactions (top row)

Since Class is a binary variable, all points lie on either 0 or 1, forming two horizontal bands.

Observation: The scatter plot shows that both legitimate (Class = 0) and fraudulent (Class = 1) transactions occur throughout the entire 48-hour observation period. Fraudulent transactions are not concentrated within a specific time interval, suggesting that fraud can occur at any point during the data collection period. The plot also highlights the severe class imbalance, with legitimate transactions greatly outnumbering fraudulent ones. Based on this visualization, the Time feature alone does not appear to clearly distinguish fraudulent transactions from legitimate ones.

Amount vs Time (Shows spending behaviour)

```python
plt.figure(figsize=(10,6))

plt.scatter(
    df["Time"],
    df["Amount"],
    alpha=0.3,
    s=10
)

plt.title("Transaction Amount over Time")
plt.xlabel("Time")
plt.ylabel("Amount")

plt.show()
```

Observation: The scatter plot indicates that most transactions involve relatively small amounts, with a dense concentration of observations below approximately 1,000. A small number of high-value transactions are observed throughout the 48-hour period, indicating the presence of outliers. There is no apparent linear relationship between transaction amount and time, as transaction amounts remain widely dispersed across the entire observation period. This suggests that transaction amount does not systematically increase or decrease over time.
